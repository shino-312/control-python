```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct
import control.matlab as ctm
```

```python
# 呼び出すたびに異なる線種を返す
def linestyle_generator():
    styles = ['-', '--', '-.', ':']
    i = 0
    while True:
        yield styles[i]
        i = (i+1) % len(styles)

# グラフの見た目を設定する
def plot_set(fig_ax, *args):
    fig_ax.set_xlabel(args[0])
    fig_ax.set_ylabel(args[1])
    fig_ax.grid(ls=':')
    if len(args) == 3:
        fig_ax.legend(loc=args[2])

# ボード線図の見た目を設定する
def bodeplot_set(fig_ax, *args):
    fig_ax[0].grid(which='both', ls=':')
    fig_ax[0].set_ylabel('Gain [dB]')
    fig_ax[1].grid(which='both', ls=':')
    fig_ax[1].set_xlabel('$\omega$ [rad/s]')
    fig_ax[1].set_ylabel('Phase [deg]')
    
    if len(args) > 0:
        fig_ax[0].legend(loc=args[0])
    if len(args) > 1:
        fig_ax[1].legend(loc=args[1])
```

```python
# Change T(time constant:時定数)
LS = linestyle_generator()
K = 1
T = [0.1, 0.5, 1.0]

fig, ax = plt.subplots()

for Ti in T:
    P = ct.tf([K], [Ti,1])
    y,t = ctm.step(P, np.arange(0, 5, 0.01))
    ax.plot(t,y,label='T='+str(Ti),ls=next(LS))

plot_set(ax,'t','y','best')
plt.show()
```

```python
# Change K(gain)
LS = linestyle_generator()
K = [1, 2, 3]
T = 0.5

fig, ax = plt.subplots()

for Ki in K:
    P = ct.tf([Ki], [T,1])
    y,t = ctm.step(P, np.arange(0, 5, 0.01))
    ax.plot(t,y,label='K='+str(Ki),ls=next(LS))

plot_set(ax,'t','y','best')
plt.show()
```
